# Overfitting
In order to avoid overfitting, we have to reduce the model complexity--reducing the number of parameters, restricting the value of paramters

# Weight decay

```python
%matplotlib inline
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# generate dataset
n_train, n_test = 20, 100 # less training sample-->more likely to be overfitting
num_inputs = 200
batch_size = 8

true_w = torch.randn((num_inputs, 1))*0.01
true_b = 0.05

train_data = d2l.synthetic_data(true_w, true_b, n_train)
train_iter = d2l.load_array(train_data, batch_size)

test_data = d2l.synthetic_data(true_w, true_b, n_test)
test_iter = d2l.load_array(test_data, batch_size, is_train=False)
```

```python
def init_params():
    w = torch.normal(0, 1, size=(num_inputs, 1), requires_grad=True)
    b = torch.zeros(1, requires_grad=True)
    return w, b

def L2_penalty(W):
    return torch.sum(W.pow(2))/2
def L1_penalty(W):
    return torch.sum(torch.abs(W))
```

# Here is L1 norm

```python
def train(lamba, num_epochs): # you cannot write lambda
    w, b = init_params()
    net, loss = lambda X:d2l.linreg(X, w, b), d2l.squared_loss
    lr = 0.01
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log', xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter:
#             with torch.enable_grad():
            l = loss(net(X), y) + lamba*L1_penalty(w)
            # update
            l.sum().backward()
            d2l.sgd([w,b], lr, batch_size)
        if (epoch+1)%5==0:
            animator.add(epoch+1, (d2l.evaluate_loss(net, train_iter, loss), d2l.evaluate_loss(net, test_iter, loss)))
    print('L2 norm of w: ', torch.norm(w).item())    
```

```python
train(0, 20)
```

```python
train(0, 200)
```

```python
train(1, 20)
```

```python
train(1, 200)
```

Based on the experiment:
- regression with L1 norm will converge faster but not stable.

# Overfitting is kind of be depressed.

# Wight decay with pytorch

```python
def train_concise(wd):
    model = nn.Sequential(nn.Linear(num_inputs, 1))
    # init parametersabs
    for param in model.parameters():
        param.data.normal_()
    loss = nn.MSELoss()
    num_epochs, lr = 100, 0.003
    
    optimizer = torch.optim.SGD(
        [{"params": model[0].weight, 'weight_decay': wd}, {"params": model[0].bias}],
        lr=lr)
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log',
                            xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter:
            with torch.enable_grad():
                optimizer.zero_grad()
                l = loss(model(X), y)
            l.backward()
            optimizer.step()
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1, (d2l.evaluate_loss(model, train_iter, loss),
                                     d2l.evaluate_loss(model, test_iter, loss)))
    print('L2 norm of w: ', model[0].weight.norm().item())
```

```python
train_concise(1)
```
